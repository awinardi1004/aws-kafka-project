```python
import websocket
import json
from kafka import KafkaProducer
```

```python
API_KEY = "****"
```

```python
# Konfigurasi Kafka
producer = KafkaProducer(
    bootstrap_servers=[":9092"],
    value_serializer=lambda v: json.dumps(v).encode("utf-8")
)
```

```python
def send_to_kafka(message):
    producer.send("stock_stream", message)
    producer.flush()
```

```python
def on_message(ws, message):
    msg = json.loads(message)

    # Hanya kirim tipe pesan "trade"
    if msg.get("type") == "trade":
        for trade in msg["data"]:
            send_to_kafka(trade)
            print("Sent to Kafka:", trade)
```

```python
def on_open(ws):
    print("Connected to Finnhub WebSocket")
    symbols = ["AAPL", "GOOGL", "TSLA", "MSFT"]

    for sym in symbols:
        ws.send(json.dumps({"type": "subscribe", "symbol": sym}))
        print("Subscribed:", sym)
```

```python
# Untuk Notebook, jalankan di cell ini
socket_url = f"wss://ws.finnhub.io?token={API_KEY}"
ws = websocket.WebSocketApp(
    socket_url,
    on_message=on_message,
    on_open=on_open
)

ws.run_forever()
```
